## 2.1　相関の強さ

- URL : http://kogolab.chillout.jp/elearn/icecream/chap2/sec1.html
- テーマ
   1. matplotlibを使う
       - 2枚以上のグラフを同時に表示する(subplot)
       - グラフのサイズを変える
       - グラフのタイトルも追加してみる

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math

# csvからデータの読み込み
df_1_1 = pd.read_csv('./data/chapter2-1_1.csv')
df_1_2 = pd.read_csv('./data/chapter2-1_2.csv')

# グラフ表示
# グラフのサイズ変更
plt.figure(figsize=(12, 4))

# subplot 1 (1行/2列で配置、1番目のグラフ設定)
# X,Y軸のスケールはグラフを比較しやすくするため同じにする
plt.subplot(1, 2, 1)
plt.title("case 1")
plt.xlim([25, 35])
plt.ylim([200, 500])
plt.xlabel('high temp[degrees Celsius]')
plt.ylabel('guests[num]')
plt.grid()
plt.scatter(df_1_1.high_temp, df_1_1.guests, marker="o", color="red")

# subplot 1 (1行/2列で配置、2番目のグラフ設定)
plt.subplot(1, 2, 2)
plt.title("case 2")
plt.xlim([25, 35])
plt.ylim([200, 500])
plt.xlabel('high temp[degrees Celsius]')
plt.ylabel('guests[num]')
plt.grid()
plt.scatter(df_1_2.high_temp, df_1_2.guests, marker="x", color="blue")
```

## 2.2　相関係数

- URL : http://kogolab.chillout.jp/elearn/icecream/chap2/sec2.html
- テーマ
   1. 相関係数関連の機能を使う
       - 散布図行列を描く
       - pandasの関数で相関係数を求める

```python
# 散布図行列を書く
pd.tools.plotting.scatter_matrix(df_1_1)
```

```python
# 散布図行列を書く
# subplotでは2つの散布図行列を並べて表示が出来なかった・・・
pd.tools.plotting.scatter_matrix(df_1_2)
```

```python
# 相関係数を求める：pandasの関数を使用
df_1_1.corr()
```

```python
# 必要な要素にアクセス(guests vs high_temp の相関関数のみ取り出す)
corr_1_2 = df_1_2.corr()
corr_1_2.guests.high_temp
```
