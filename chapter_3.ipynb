```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math

# csvからデータの読み込み
df_1 = pd.read_csv('./data/chapter3-2.csv')

# 散布図行列を書く
pd.tools.plotting.scatter_matrix(df_1)
```

```python
# 相関係数を求める(p値も見たいので、scipyの関数を使用)
corr_list_1 = stats.pearsonr(df_1.age, df_1.store_visitor)
print('r_val = ', corr_list_1[0])
print('p_val = ', corr_list_1[1])
```

```python
# csvからデータの読み込み
df_2 = pd.read_csv('./data/chapter3-6.csv')

# 相関係数・p値を求める
corr_list_2 = stats.pearsonr(df_2.score, df_2.grade)
print('r_val = ', corr_list_2[0])
print('p_val = ', corr_list_2[1])
```

```python
# 散布図行列を書く
pd.tools.plotting.scatter_matrix(df_2)
```

### 上の結果からわかること
- 帰無仮説
    - 入学試験の点数と学業成績に相関は無い
- 対立仮説
    - 入学試験の点数と学業成績に相関は有る
- 有意水準
    - 1%とする
- p値
    - 0.00015
- 検定結果
    - p値が有意水準を下回っている。このため帰無仮説を棄却し、対立仮説を採択する。
    - つまり、入学試験の点数と学業成績には有意な正の相関がある。
        - 入学試験の点数が高い人ほど学業成績も高い
