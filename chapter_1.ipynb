## 1.2　散布図を描く

- URL : http://kogolab.chillout.jp/elearn/icecream/chap1/sec2.html
- テーマ
   1. matplotlibを使用する
       - 折れ線グラフを作成する
       - 散布図を作成する
       - マーカーの形や色を指定してみる

```python
plt.xlim([0, 15])
plt.ylim([100, 800])
plt.xlabel('data[day]')
plt.ylabel('guests[num]')
plt.grid()
plt.plot(df_1.date, df_1.guests, marker="o", color="red")
```

```python
plt.xlim([25, 35])
plt.ylim([200, 700])
plt.xlabel('high temp[degrees Celsius]')
plt.ylabel('guests[num]')
plt.grid()
plt.scatter(df_1.high_temp, df_1.guests, marker="x", color="blue")
```
