## 7.6　通過テスト

- URL : http://kogolab.chillout.jp/elearn/icecream/chap7/sec6.html
- テーマ
   1. 相関係数の抽出
       - ある一定のルールに従って、相関係数を抽出する。
           - 正の相関が高いもの（ここでは+0.4以上）
           - 負の相関が高いもの（ここでは-0.2以下）

```python
# データフレームの表の書式を変える
# 相関係数の値によって、以下のように表示する
#  -0.2以下 : 赤太字
#  0.4以上 : 青太字

df_2 = df_1.corr()

def highlight_value(val):
    if val == 1.0:
        return 'color: {0}'.format('black')
    elif val <= -0.2:
        return 'color: {0}; font-weight: bold'.format('red')
    elif val >= 0.4:
        return 'color: {0}; font-weight: bold'.format('blue')
    else:
        return 'color: {0}'.format('black')

df_2.style.applymap(highlight_value)
```

```python
# データフレームの中を行/列でfor文で回す
# 行名と列名が同じものは相関関数=1なので除外する
# 以下の2条件に当てはまる値のみを表示する
# 1)正の相関が高いもの（ここでは+0.4以上）
# 2)負の相関が高いもの（ここでは-0.2以下）

for var1 in range(0, df_2.shape[0]-1):
    for var2 in range(0, df_2.shape[1]-1):
        df_split = df_2.ix[var1:var1+1, var2:var2+1]
        c_name = df_split.columns.values
        r_name = df_split.index.values
        value  = df_split.values
        if(c_name != r_name) and ((value <= -0.2) or (value >= 0.4)):
        #if(c_name != r_name) and (value <= -0.2):
            print(c_name, r_name, value)
```
